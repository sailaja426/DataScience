```python
import yfinance as yf
import pandas as pd

# Create the ticker object for Tesla
ticker = yf.Ticker("TSLA")

# Retrieve stock information for Tesla with maximum available period
tesla_revenue = ticker.history(period="max")

# Reset the index of the dataframe
tesla_revenue.reset_index(inplace=True)

# Display the first five rows of the dataframe
print(tesla_revenue.head())
```

```python

tesla_revenue.tail()
```

```python
# Create the ticker object for Tesla
ticker = yf.Ticker("GME")

# Retrieve stock information for GME with maximum available period
gme_revenue = ticker.history(period="max")

# Reset the index of the dataframe
gme_revenue.reset_index(inplace=True)

# Display the first five rows of the dataframe
print(gme_revenue.head())
```

```python
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

def make_graph(dataframe):
    plt.figure(figsize=(12, 6))
    plt.plot(dataframe['Date'], dataframe['Close'])
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title('Tesla Stock Price Over Time')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
    
make_graph(tesla_revenue)
```
